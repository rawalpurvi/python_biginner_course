## Games

### Hangman

```python
import random


HANGMAN = [
    '________',
    '|       |',
    '|       O',
    '|       |',
    '|      /|\ ',
    '|       |',
    '|      / \ '
]

WORDS = [
    'JAVA', 'PERL', 'RUBY', 'SQL', 'PYTHON', 'DJANGO',
    'BASIC', 'COBOL', 'JAVASCRIPT', 'ASSEMBLYLANGUAGE', 'UNITTESTING'
]

class Hangman:
    
    """
    The hangman game class with his methods
    """

    def __init__(self,guess_word):
        
        self.user_guess_word = list("_" * len(guess_word))
        self.wrong_guess = 0
        self.guess_word = guess_word
        
        
    """
    Method get_input  : return input value form user
    
    """

    def get_input(self):
        
        # Get input from user to guess the word
        
        guess_char = input("Give one character to guess the word : ")
        return guess_char.upper()
    
    
    """
    Method is_valid_input  : return True of False 
    
    """
    
    def is_valid_input(self,guess_char):
        
        # Check user input is correct or not
        
        return (guess_char.isalpha() and len(guess_char) == 1)
        
    """
    Method change_user_guess_word_list  : return updated user_guess_word list
    
    """
    
    def change_user_guess_word_list(self,guess_char):
        
        # Add guess word in user_guess_word list
    
        for i , ch in enumerate(self.guess_word):
            if ch == guess_char :
                self.user_guess_word[i] = ch
                
    """
    Method print_user_guess  : print hangman and user_guess_word list
    
    """        
    
    def print_user_guess(self):
        
        # Display user guess word
        
        for i in range(0,self.wrong_guess):
            print(HANGMAN[i])
        print("\n")
        print(' '.join(self.user_guess_word))
        print("\n")
        
    
    """
    Method play_game  : play hangman game logic
    
    """  
    
    def play_game(self):

        # user can give wrong input till hangman list for display
        while self.wrong_guess < len(HANGMAN):

            # Print hangman and the guess word
            self.print_user_guess()

            # Get input from user
            guess_char = self.get_input()
                       
            # Check if user input is valid or not    
            if self.is_valid_input(guess_char):
                
                #if user already guess the alphabet before
                
                if guess_char in self.user_guess_word:
                    print("You already guess that alphabet")
                    continue
                
                # Check if uesr input is in guess word
                if guess_char in self.guess_word:
                    self.change_user_guess_word_list(guess_char)
                else:
                    self.wrong_guess += 1
            else:
                print("Please give correct input")
                continue
                          
            # Check if there is no _ in user_guess_word list
            if "_" not in self.user_guess_word:
                print(f"Yoo are a winner!!\n\n The word is {self.guess_word}")
                break
        else:
            print("You loose!!")
            
            
# Play Hangman Game            
guess_word = random.choice(WORDS)
game_hangman = Hangman(guess_word)
game_hangman.play_game()        
```
